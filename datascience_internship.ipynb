```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
np.random.seed(42) # for reproducibility
data = np.random.rand(100, 6) # 100 rows, 6 columns
df = pd.DataFrame(data, columns=[f'Column_{i}' for i in range(6)])

# Introduce some random missing values
for col in df.columns:
    df.loc[df.sample(frac=0.1).index, col] = np.nan # introduce 10% missing values

print("Original DataFrame:")
display(df.head()) # Displaying head as the DataFrame is large

# Imputation using mean
df_mean_imputed = df.fillna(df.mean())
print("\nDataFrame after mean imputation:")
display(df_mean_imputed.head())

# Imputation using median
df_median_imputed = df.fillna(df.median())
print("\nDataFrame after median imputation:")
display(df_median_imputed.head())

# Imputation using forward fill (ffill())
df_ffill_imputed = df.ffill()
print("\nDataFrame after forward fill imputation:")
display(df_ffill_imputed.head())

# Imputation using backward fill (bfill())
df_bfill_imputed = df.bfill()
print("\nDataFrame after backward fill imputation:")
display(df_bfill_imputed.head())

# Imputation using a constant value (e.g., 0)
df_constant_imputed = df.fillna(0)
print("\nDataFrame after constant imputation (0):")
display(df_constant_imputed.head())
```

## Statistical outlier detection

### Subtask:
Explain and implement a common statistical method (e.g., Z-score or IQR) for outlier detection.


**Reasoning**:
Calculate the IQR, determine the lower and upper bounds, and identify the outliers in the 'Numerical_Column' of the `df_outliers` DataFrame.



```python
# Calculate Q1 and Q3
Q1 = df_outliers['Numerical_Column'].quantile(0.25)
Q3 = df_outliers['Numerical_Column'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Determine the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = df_outliers[(df_outliers['Numerical_Column'] < lower_bound) | (df_outliers['Numerical_Column'] > upper_bound)]

# Print the identified outliers
print("\nIdentified Outliers using IQR method:")
display(outliers)
```

## Feature selection (filtering)

### Subtask:
Demonstrate a filtering method for feature selection.


**Reasoning**:
Calculate the correlation matrix, extract correlations with the target, sort them, select the top N features, and print the selected feature names to demonstrate a filtering method for feature selection.



```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Extract the absolute correlations with the 'Target' variable
target_correlation = correlation_matrix['Target'].abs()

# Sort the features based on their absolute correlation with the 'Target'
sorted_correlations = target_correlation.sort_values(ascending=False)

# Select the top N features (excluding the 'Target' itself)
N = 5
top_features = sorted_correlations.drop('Target').head(N)

# Print the names of the selected features
print(f"Top {N} features based on correlation with Target:")
print(top_features)
```

## Feature selection (wrapper)

### Subtask:
Demonstrate a wrapper method for feature selection.


**Reasoning**:
Implement Recursive Feature Elimination (RFE) with a Linear Regression model to select the top 5 features.



```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

# Define features (X) and target (y)
X = df.drop('Target', axis=1)
y = df['Target']

# Initialize the base estimator
estimator = LinearRegression()

# Initialize RFE with the estimator and desired number of features
# RFE requires numerical data, so we'll drop the original categorical column if it still exists
if 'Categorical_Feature' in X.columns:
    X = X.drop('Categorical_Feature', axis=1)

# Ensure all feature columns are numeric for RFE
X = X.select_dtypes(include=np.number)


n_features_to_select = 5
rfe = RFE(estimator=estimator, n_features_to_select=n_features_to_select)

# Fit RFE to the data
rfe.fit(X, y)

# Get the ranking and support (boolean mask) of features
feature_ranking = rfe.ranking_
selected_features_mask = rfe.support_

# Get the names of the selected features
selected_features = X.columns[selected_features_mask]

# Print the selected features
print("Selected features using RFE:")
print(selected_features)
```

## Feature selection (embedded)

### Subtask:
Demonstrate an embedded method for feature selection.


**Reasoning**:
Demonstrate an embedded method for feature selection using Lasso regression and SelectFromModel.



```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# Define features (X) and target (y)
# Using the standard scaled data as Lasso is sensitive to scale
X = df_standard_scaled.drop('Target', axis=1)
y = df_standard_scaled['Target']

# Initialize the Lasso model
# A smaller alpha value means less regularization and might select more features
lasso = Lasso(alpha=0.01, random_state=42)

# Initialize SelectFromModel with the Lasso model
# It will select features whose coefficients are above the threshold (default is 1e-5 for Lasso)
sfm = SelectFromModel(estimator=lasso)

# Fit SelectFromModel to the data
sfm.fit(X, y)

# Get the boolean mask of the selected features
selected_features_mask_embedded = sfm.get_support()

# Get the names of the selected features
selected_features_embedded = X.columns[selected_features_mask_embedded]

# Print the selected features
print("Selected features using Embedded method (Lasso):")
print(selected_features_embedded)
```

## Inferential statistics (confidence interval)

### Subtask:
Calculate a confidence interval for the mean of the numerical column to demonstrate inferential statistics.


**Reasoning**:
Import norm from scipy.stats, calculate the sample mean, standard error, confidence interval, and print the result.



```python
from scipy.stats import norm
import numpy as np # Import numpy as it is used for np.sqrt
import pandas as pd

# Create a dictionary with random data
data = {
    'Numerical_Feature': np.random.randn(100),
    'Categorical_Feature': np.random.choice(['Group A', 'Group B', 'Group C'], 100)
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Calculate the sample mean
sample_mean = df['Numerical_Feature'].mean()

# Calculate the standard error of the mean
sample_std = df['Numerical_Feature'].std()
sample_size = len(df)
standard_error = sample_std / np.sqrt(sample_size)

# Define the confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the z-score
z_score = norm.ppf(1 - alpha/2)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

# Print the calculated confidence interval
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Standard Error: {standard_error:.4f}")
print(f"Confidence Interval ({confidence_level*100}%): [{confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}]")
```

## Hypothesis testing (t-test)

### Subtask:
Perform a hypothesis test (e.g., an independent samples t-test if you have two groups in your categorical data) to compare the means of the numerical feature between different categories.


**Reasoning**:
Perform an independent samples t-test on the numerical feature for two different categories to compare their means.



```python
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd

# Create a dictionary with random data
data = {
    'Numerical_Feature': np.random.randn(100),
    'Categorical_Feature': np.random.choice(['Group A', 'Group B', 'Group C'], 100)
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Identify unique categories
# Use the DataFrame from the data creation step of this task
unique_categories = df['Categorical_Feature'].unique()
print(f"Unique categories found: {unique_categories}")

# Check if there are at least two unique categories
if len(unique_categories) >= 2:
    # Select two categories for the t-test
    # Ensure the selected categories exist in the unique_categories list
    category_1 = 'Group A' # Assuming 'Group A' is one of the unique categories
    category_2 = 'Group B' # Assuming 'Group B' is another unique category

    if category_1 in unique_categories and category_2 in unique_categories:
        # Filter data for each category
        data_category_1 = df[df['Categorical_Feature'] == category_1]['Numerical_Feature']
        data_category_2 = df[df['Categorical_Feature'] == category_2]['Numerical_Feature']

        # Perform independent samples t-test
        # Check if there are enough samples in each group
        if len(data_category_1) > 1 and len(data_category_2) > 1:
            t_statistic, p_value = ttest_ind(data_category_1, data_category_2)

            # Print the results
            print(f"\nIndependent Samples t-test between '{category_1}' and '{category_2}':")
            print(f"T-statistic: {t_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")

            # Interpret the result (common significance level alpha = 0.05)
            alpha = 0.05
            if p_value < alpha:
                print(f"Result: Reject the null hypothesis. There is a statistically significant difference in means between '{category_1}' and '{category_2}'.")
            else:
                print(f"Result: Fail to reject the null hypothesis. There is no statistically significant difference in means between '{category_1}' and '{category_2}'.")
        else:
            print(f"\nNot enough data points in one or both categories ('{category_1}', '{category_2}') to perform an independent samples t-test.")

    else:
        print(f"\nSelected categories ('{category_1}', '{category_2}') not found in the unique categories of 'Categorical_Feature'.")

else:
    print("\nNot enough unique categories in 'Categorical_Feature' to perform an independent samples t-test.")
```
