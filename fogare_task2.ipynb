## Handle missing values

### Subtask:
Address the missing values in the `Income`, `Credit_Score`, and `Loan_Balance` columns.


**Reasoning**:
Impute the missing values in 'Income', 'Credit_Score', and 'Loan_Balance' using the median and verify the imputation.



```python
for col in ['Income', 'Credit_Score', 'Loan_Balance']:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

display(df.isnull().sum())
```

## Calculate correlation

### Subtask:
Calculate and visualize the correlation matrix for the numerical columns to understand the relationships between them.


**Reasoning**:
Select numerical columns, calculate the correlation matrix, and visualize it using a heatmap with annotations.



```python
numerical_cols = df.select_dtypes(include=['number'])
correlation_matrix = numerical_cols.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

## Formulate and test hypotheses

### Subtask:
Formulate and test hypotheses based on the data and potential areas of interest. This will require understanding what specific relationships or differences you are interested in investigating.


**Reasoning**:
Identify and formulate hypotheses based on the data, and outline the approach for testing them.



```python
# Hypothesis 1: Is there a significant difference in Income based on Employment Status?
# Null Hypothesis (H0_1): The mean income is the same across different employment statuses.
# Alternative Hypothesis (H1_1): The mean income is significantly different across different employment statuses.
# Test: ANOVA (Analysis of Variance) because we are comparing the means of a numerical variable (Income) across more than two groups (Employment_Status).

# Data Preparation for Hypothesis 1:
# Group the data by 'Employment_Status' and extract 'Income' for each group.

# Statistical Test for Hypothesis 1:
# Use scipy.stats.f_oneway to perform the ANOVA test.

# Interpretation of Results for Hypothesis 1:
# If the p-value is less than the significance level (e.g., 0.05), reject H0_1 and conclude that there is a significant difference in mean income based on employment status.
# If the p-value is greater than or equal to the significance level, fail to reject H0_1.

# Hypothesis 2: Is there a significant correlation between Credit_Score and Loan_Balance?
# Null Hypothesis (H0_2): There is no significant correlation between Credit_Score and Loan_Balance.
# Alternative Hypothesis (H1_2): There is a significant correlation between Credit_Score and Loan_Balance.
# Test: Pearson correlation test because we are testing the linear relationship between two numerical variables.

# Data Preparation for Hypothesis 2:
# Select the 'Credit_Score' and 'Loan_Balance' columns.

# Statistical Test for Hypothesis 2:
# Use scipy.stats.pearsonr to perform the Pearson correlation test.

# Interpretation of Results for Hypothesis 2:
# If the p-value is less than the significance level (e.g., 0.05), reject H0_2 and conclude that there is a significant correlation between Credit_Score and Loan_Balance.
# If the p-value is greater than or equal to the significance level, fail to reject H0_2.

print("Hypotheses formulated and test approaches outlined.")
```

**Reasoning**:
Perform the ANOVA test to examine the relationship between Income and Employment Status.



```python
from scipy import stats

# Prepare data for Hypothesis 1 (ANOVA)
employment_statuses = df['Employment_Status'].unique()
income_groups = [df['Income'][df['Employment_Status'] == status].dropna() for status in employment_statuses]

# Perform ANOVA test
f_statistic, p_value_h1 = stats.f_oneway(*income_groups)

print(f"ANOVA Test Results for Income vs. Employment Status:")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value_h1:.4f}")

# Interpret the results
alpha = 0.05
if p_value_h1 < alpha:
    print(f"Result: Reject the null hypothesis. There is a significant difference in mean income based on employment status.")
else:
    print(f"Result: Fail to reject the null hypothesis. There is no significant difference in mean income based on employment status.")
```

**Reasoning**:
Perform the Pearson correlation test to examine the relationship between Credit_Score and Loan_Balance.



```python
# Prepare data for Hypothesis 2 (Pearson correlation)
credit_scores = df['Credit_Score'].dropna()
loan_balances = df['Loan_Balance'].dropna()

# Perform Pearson correlation test
correlation_coefficient, p_value_h2 = stats.pearsonr(credit_scores, loan_balances)

print(f"\nPearson Correlation Test Results for Credit_Score vs. Loan_Balance:")
print(f"Correlation Coefficient: {correlation_coefficient:.4f}")
print(f"P-value: {p_value_h2:.4f}")

# Interpret the results
alpha = 0.05
if p_value_h2 < alpha:
    print(f"Result: Reject the null hypothesis. There is a significant correlation between Credit_Score and Loan_Balance.")
else:
    print(f"Result: Fail to reject the null hypothesis. There is no significant correlation between Credit_Score and Loan_Balance.")
```

## Explore other aspects

### Subtask:
Explore other aspects of the data, such as analyzing categorical variables, creating visualizations, or identifying patterns.


**Reasoning**:
Analyze the distribution of categorical variables and visualize them using bar plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = ['Employment_Status', 'Credit_Card_Type', 'Location']
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()

monthly_payment_cols = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']
for col in monthly_payment_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col} Payment Status')
    plt.xlabel('Count')
    plt.ylabel('Payment Status')
    plt.show()
```

**Reasoning**:
Visualize the relationship between numerical and categorical variables using box plots.



```python
numerical_cols = ['Income', 'Credit_Score', 'Loan_Balance']
categorical_cols = ['Employment_Status', 'Credit_Card_Type']

for num_col in numerical_cols:
    for cat_col in categorical_cols:
        plt.figure(figsize=(12, 7))
        sns.boxplot(data=df, x=cat_col, y=num_col, palette='pastel')
        plt.title(f'{num_col} Distribution by {cat_col}')
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

**Reasoning**:
Explore patterns in the monthly payment status columns and visualize the proportion of each payment status over time using a stacked bar chart.



```python
monthly_payment_cols = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']
monthly_payment_counts = df[monthly_payment_cols].apply(pd.Series.value_counts).transpose()

monthly_payment_counts.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')
plt.title('Monthly Payment Status Distribution Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.legend(title='Payment Status')
plt.tight_layout()
plt.show()
```
