## Identify missing values

### Subtask:
Check for missing values in each column and visualize the missing data distribution.


**Reasoning**:
Calculate the number of missing values per column and visualize the missing data distribution using missingno.



```python
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)

import missingno as msno
msno.matrix(df)
```

## Analyze missing value patterns

### Subtask:
Determine which columns have missing values and understand the extent of missingness.


**Reasoning**:
Calculate and display the percentage of missing values for columns with missing data.



```python
total_rows = len(df)
columns_with_missing_values = missing_values[missing_values > 0]

print("\nColumns with missing values and their percentages:")
for column, count in columns_with_missing_values.items():
    percentage = (count / total_rows) * 100
    print(f"{column}: {count} missing values ({percentage:.2f}%)")
```

## Impute missing values

### Subtask:
Based on the data type of columns with missing values, impute them using appropriate central tendency measures (mean for numerical, median for skewed numerical, mode for categorical).


**Reasoning**:
Identify the data types of the columns with missing values and then decide whether to use mean or median for imputation based on the distribution of the numerical columns. Finally, impute the missing values using the selected central tendency measures.



```python
print("Data types of columns with missing values:")
print(df[columns_with_missing_values.index].dtypes)

# Check distribution of numerical columns with missing values
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols_with_missing = df[columns_with_missing_values.index].select_dtypes(include=['number']).columns

for col in numerical_cols_with_missing:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Impute missing values
for col in columns_with_missing_values.index:
    if df[col].dtype in ['int64', 'float64']:
        # Decide whether to use mean or median based on distribution (visually from the plots)
        # For demonstration, let's use median for potentially skewed data like Income and Loan_Balance, and mean for Credit_Score
        if col in ['Income', 'Loan_Balance']:
            imputation_value = df[col].median()
            print(f"Imputing missing values in '{col}' with median: {imputation_value}")
        else: # Credit_Score
            imputation_value = df[col].mean()
            print(f"Imputing missing values in '{col}' with mean: {imputation_value}")
        df[col].fillna(imputation_value, inplace=True)
    else:
        # For categorical columns, use mode (although there are no categorical columns with missing values in this case)
        imputation_value = df[col].mode()[0]
        print(f"Imputing missing values in '{col}' with mode: {imputation_value}")
        df[col].fillna(imputation_value, inplace=True)

print("\nMissing values after imputation:")
print(df.isnull().sum())

display(df.head())
```

## Verify imputation

### Subtask:
Verify if there are any remaining missing values after imputation.


**Reasoning**:
Check for missing values across the entire DataFrame after imputation to confirm successful handling.



```python
missing_values_after_imputation = df.isnull().sum()
print("Missing values per column after imputation:")
print(missing_values_after_imputation)
```

*   Visual inspection of the distribution of numerical columns with missing values ('Income', 'Credit\_Score', and 'Loan\_Balance') was performed using histograms.
*   Missing values were imputed using central tendency measures: the median (107658.0) for 'Income', the mean (577.72) for 'Credit\_Score', and the median (45776.0) for 'Loan\_Balance'.
*   Verification after imputation confirmed that all missing values in the DataFrame were successfully handled, with each column showing 0 missing values.

### Insights or Next Steps

## Identify outliers

### Subtask:
Detect outliers in numerical columns that could affect model performance.


**Reasoning**:
Identify numerical columns, calculate IQR and bounds for each, count outliers, and print the counts.



```python
numerical_cols = df.select_dtypes(include=['number']).columns

outliers_count = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_count[col] = outliers.shape[0]

print("Number of outliers per numerical column (using IQR):")
for col, count in outliers_count.items():
    print(f"{col}: {count} outliers")
```

## Analyze categorical features and delinquency

### Subtask:
Explore the relationship between categorical features ('Employment_Status', 'Credit_Card_Type', 'Location', and 'Month_X' payment statuses) and delinquency using visualizations and/or statistical tests.


**Reasoning**:
Create a list of categorical columns and then iterate through this list to generate count plots, cross-tabulations, calculate proportions of delinquent accounts, and create bar plots for each categorical column.



```python
categorical_cols = ['Employment_Status', 'Credit_Card_Type', 'Location', 'Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']

for col in categorical_cols:
    # Count plot for the categorical column
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()

    # Cross-tabulation with Delinquent_Account
    crosstab_result = pd.crosstab(df[col], df['Delinquent_Account'])
    print(f"\nCross-tabulation for {col} and Delinquent_Account:\n")
    print(crosstab_result)

    # Calculate the proportion of delinquent accounts within each category
    crosstab_proportion = crosstab_result.div(crosstab_result.sum(axis=1), axis=0)

    # Bar plot for the proportion of delinquent accounts
    plt.figure(figsize=(10, 6))
    crosstab_proportion[1].sort_values(ascending=False).plot(kind='barh')
    plt.title(f'Proportion of Delinquent Accounts by {col}')
    plt.xlabel('Proportion of Delinquent Accounts (1)')
    plt.ylabel(col)
    plt.show()
```
